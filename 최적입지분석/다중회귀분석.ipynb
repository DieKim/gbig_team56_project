```python
nigs.head()
```

```python
nigs_m = pd.DataFrame(nigs['사업장명'].groupby(nigs['행정동']).count())
nigs_m.reset_index(inplace=True)
nigs_m.columns=['행정동','노인교실수']
nigs_m
```

```python
old_ratio.rename(columns={'동':'행정동'},inplace=True)
old_ratio.head()
```

```python
old_act=old_act[['행정동','시간대구분','60세이상평균']]

old_act_dawn_df =old_act[old_act['시간대구분']=='새벽']
old_act_dawn_df.reset_index(inplace=True,drop=True)

old_act_morning_df =old_act[old_act['시간대구분']=='아침']
old_act_morning_df.reset_index(inplace=True,drop=True)

old_act_day_df =old_act[old_act['시간대구분']=='낮']
old_act_day_df.reset_index(inplace=True,drop=True)

old_act_afternoon_df =old_act[old_act['시간대구분']=='저녁']
old_act_afternoon_df.reset_index(inplace=True,drop=True)

old_act_night_df =old_act[old_act['시간대구분']=='밤']
old_act_night_df.reset_index(inplace=True,drop=True)
```

```python
old_act_dawn = pd.DataFrame(old_act_dawn_df['60세이상평균'].groupby(old_act_dawn_df['행정동']).mean())
old_act_dawn.reset_index(inplace=True)

old_act_morning = pd.DataFrame(old_act_morning_df['60세이상평균'].groupby(old_act_morning_df['행정동']).mean())
old_act_morning.reset_index(inplace=True)

old_act_day = pd.DataFrame(old_act_day_df['60세이상평균'].groupby(old_act_day_df['행정동']).mean())
old_act_day.reset_index(inplace=True)

old_act_afternoon = pd.DataFrame(old_act_afternoon_df['60세이상평균'].groupby(old_act_afternoon_df['행정동']).mean())
old_act_afternoon.reset_index(inplace=True)

old_act_night = pd.DataFrame(old_act_night_df['60세이상평균'].groupby(old_act_night_df['행정동']).mean())
old_act_night.reset_index(inplace=True)
```

```python
# 2) vif로 다중공선성 확인
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(x_data.values, i) for i in range(x_data.shape[1])]
vif["features"] = x_data.columns
vif
```
