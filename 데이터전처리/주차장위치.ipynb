행정동목록데이터불러오기

```python
with open('/content/drive/MyDrive/서울행정동.csv', 'r',encoding = 'cp949') as f:
   reader = csv.reader(f)
   dong = list(reader)
```

```python
#행정동 개수
len(dong)
```

```python
def getLatLng():
    max_data = 15
    data = 0
    id = []
    name = []
    address = []
    lon = []
    lat = []

    result = []


    for i in range(0, len(dong)):
        number = 0
        searching = '서울특별시 ' + str(dong[i][0]) + ' 지하철'
        url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(searching)
        header = {'Authorization': 'KakaoAK ' + APP_KEY}

        places = requests.get(url, headers = header)

        
        while (number < max_data):
            
            try:
                # print(places.json()['documents'][number])
                # id
                if (places.json()['documents'][number]['id']):
                    id.append(places.json()['documents'][number]['id'])
                else:
                    id.append('')

                # 주차장 이름
                if (places.json()['documents'][number]['place_name']):
                    name.append(places.json()['documents'][number]['place_name'])
                else:
                    name.append('')

                # 주소
                if (places.json()['documents'][number]['address_name']):
                    address.append(places.json()['documents'][number]['address_name'])
                else:
                    address.append('')

                # 경도
                if (places.json()['documents'][number]['x']):
                    lon.append(places.json()['documents'][number]['x'])
                else:
                    lon.append('')
        
                # 위도
                if (places.json()['documents'][number]['y']):
                    lat.append(places.json()['documents'][number]['y'])
                else:
                    lat.append('')
                
                result.append([id[data], name[data], address[data], lon[data], lat[data]])
                number = number + 1
                data = data + 1
                if (number == 14):
                    print(str(dong[i][0])," 15개 주차장 데이터 수집  완료",data)

            except IndexError:
                print(str(dong[i][0])," ",number,"개 주차장 데이터 수집  완료",data)
                break
    return result
```

```python
result = getLatLng()
```

```python
jucha = pd.DataFrame(result)
jucha.columns = ['주차장코드', '주차장명', '주소', '경도', '위도']
jucha
```

```python
#중복값확인
print(sum(jucha.duplicated()))
print(sum(jucha.duplicated(['주차장코드'])))
```

```python
# 중복값 제거
jucha = jucha.drop_duplicates(['주차장코드'])
jucha_2 = jucha.reset_index(drop=True)
```
